```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"CSV/stock_prices.csv")
```

```python
data.head()
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data.set_index('date',inplace=True)
```

```python
data.head()
```

```python
data.columns
```

```python
rolling_avg = data['closing_price'].rolling(window=30,min_periods=1).mean()
```

```python
rolling_avg.shape
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['closing_price'], label='Closing Price', linewidth=2)
plt.plot(rolling_avg.index, rolling_avg, label='30-Day Rolling Avg', color='orange', linestyle='--', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Prices with 30-Day Rolling Average')
plt.legend()
plt.grid(True)
plt.show()
```
